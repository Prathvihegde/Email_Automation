```python
import os
from datasets import load_dataset
from sklearn.metrics import accuracy_score, precision_recall_fscore_support
from transformers import (AutoModelForSequenceClassification, AutoTokenizer,
                          TrainingArguments, Trainer)

# Replace with the name of the LLM you're using (e.g., 'bert-base-uncased', 'gpt2')
model_name = "bert-base-uncased"

# Load the tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Function to tokenize a batch of examples
# Now includes the labels
def tokenize_function(examples):
    # Tokenize the text data and include the labels
    return tokenizer(examples["Cleaned Email Body"], padding="max_length", truncation=True,
                     return_tensors='pt')

# Define training arguments
training_args = TrainingArguments(
    output_dir="./results",
    num_train_epochs=3,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=64,
    warmup_steps=500,
    weight_decay=0.01,
    logging_dir='./logs',
    logging_steps=10,
    evaluation_strategy="epoch",
    save_strategy="epoch",
    load_best_model_at_end=True,  # Load the best model at the end of training
    metric_for_best_model="accuracy",  # Use accuracy to select the best model
)

# Update paths to your actual train and test data file locations
train_data_path = "/content/train_data.csv"
test_data_path = "/content/test_data.csv"

# Load the datasets
train_dataset = load_dataset('csv', data_files=train_data_path, split='train')
test_dataset = load_dataset('csv', data_files=test_data_path, split='train')

# Define your model (replace with your actual model)
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=2)

# Assuming your label column is named 'label'
# Rename the label column to 'labels'
train_dataset = train_dataset.rename_column("Intent", "labels")
test_dataset = test_dataset.rename_column("Intent", "labels")

# Define a function to compute metrics
def compute_metrics(pred):
    labels = pred.label_ids
    preds = pred.predictions.argmax(-1)
    precision, recall, f1, _ = precision_recall_fscore_support(labels, preds, average='weighted')
    acc = accuracy_score(labels, preds)
    return {
        'accuracy': acc,
        'f1': f1,
        'precision': precision,
        'recall': recall
    }

# Create Trainer instance
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset.map(tokenize_function, batched=True),
    eval_dataset=test_dataset.map(tokenize_function, batched=True),
    compute_metrics=compute_metrics,  # Pass the compute_metrics function
)

# Function to tokenize a batch of examples
# Now includes the labels
def tokenize_function(examples):
    # Tokenize the text data and include the labels
    return tokenizer(examples["Cleaned Email Body"], padding="max_length", truncation=True,
                     return_tensors='pt', **{'labels': examples['labels']}) # Added labels


# Evaluate the model and print metrics
eval_results = trainer.evaluate()
print(eval_results)

# Save the model
trainer.save_model("./fine_tuned_model")
```

```python
# prompt: give code to build a function that takes a new email, classifies its intent using the fine-tuned LLM, and generates its embedding using the pre-trained LLM.

!pip install sentence_transformers # Install sentence_transformers

from transformers import pipeline, AutoTokenizer # Import AutoTokenizer
from sentence_transformers import SentenceTransformer # Import SentenceTransformer

# Replace with the name of the LLM you're using (e.g., 'bert-base-uncased', 'gpt2')
model_name = "bert-base-uncased"

# Load the tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Load the fine-tuned model and tokenizer
classifier = pipeline("text-classification", model="./fine_tuned_model", tokenizer=tokenizer, device=0)

# Load the pre-trained model for embeddings (you might need to change this based on your model)
embedding_model = SentenceTransformer('all-mpnet-base-v2')

def process_email(new_email):
    """
    Classifies the intent of a new email and generates its embedding.

    Args:
        new_email: The text content of the new email.

    Returns:
        A dictionary containing the predicted intent and the email embedding.
    """
    try:
        # Classify intent
        intent_prediction = classifier(new_email)
        predicted_intent = intent_prediction[0]['label']  # Extract the predicted intent label

        # Generate embedding
        email_embedding = embedding_model.encode(new_email)

        return {"intent": predicted_intent, "embedding": email_embedding}

    except Exception as e:
        print(f"An error occurred: {e}")
        return None

# Example usage
new_email = "I need the status of my order #12345."
processed_data = process_email(new_email)

if processed_data:
    print(f"Predicted Intent: {processed_data['intent']}")
    print(f"Email Embedding: {processed_data['embedding']}")
```

```python
# prompt: final code to complete the basic retrieval system by implementing the cosine similarity calculation and the most relevant response embedding and retrieved the corresponding text for a classified email.

import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# Define example_responses here so it's accessible
example_responses = {
    "Request for Information": ["Pinnacle to proceed with the request", "Pinnacle to acknowlegde and respond", "Client to provide additional details"],
    "Acknowledgment": ["Pinnacle to proceed with final delivery", "Pinnacle to await any further feedback"],
    "Status Update": ["Client to review and give feedback"]
}

def get_most_relevant_response(email_embedding, example_responses):
    """
    Finds the most relevant response based on cosine similarity between embeddings.

    Args:
        email_embedding: The embedding of the input email.
        example_responses: A dictionary where keys are intents and values are lists of responses.

    Returns:
        The most relevant response string.
    """
    max_similarity = -1
    most_relevant_response = None

    # Generate embeddings for each example response and calculate cosine similarity
    for intent, responses in example_responses.items():
        for response in responses:
            response_embedding = embedding_model.encode(response)
            similarity = cosine_similarity([email_embedding], [response_embedding])[0][0]
            if similarity > max_similarity:
                max_similarity = similarity
                most_relevant_response = response

    return most_relevant_response


# Example Usage (assuming processed_data from the previous code cell)
if processed_data:
    # processed_data was generated using the 'process_email' function in a previous cell
    # with the email: "I need the status of my order #12345."
    example_email_embedding = processed_data["embedding"]
    most_relevant = get_most_relevant_response(example_email_embedding, example_responses)
    print(f"For the example email: 'I need the status of my order #12345.'")  # Added clarification
    print(f"Most Relevant Response: {most_relevant}")
```
